## Make a bar graph for manufacturer type.

```python
Manufacturer_counts = DF['Manufacturer'].value_counts()

plt.figure(figsize=(10, 6))
Manufacturer_counts.plot(kind='bar', color='violet')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Cars')
plt.title('Number of Cars by Manufacturer')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show() # Show the plot
```

## Make a pie chart for showing vehicle type by percentage

```python
vehicle_type_percentage = DF['Vehicle_type'].value_counts(normalize=True) * 100

# Plotting the pie chart
plt.figure(figsize=(4, 4))
plt.pie(vehicle_type_percentage, labels=vehicle_type_percentage.index, autopct='%1.1f%%',  colors=['Violet', 'red'])
plt.title('Percentage Distribution of Vehicle Types')
plt.show()
```

## Calculate average Price of all cars by using correct measure of central Tendency

```python
Average_price = DF['Price_in_thousands'].mean() #Average price of all cars after removing rows with missing values.

print(f"The average price of all cars is: ${Average_price:.2f}k") 
```

## Calculate the variance and standard deviation for numerical columns such as Sales_in_thousands, __year_resale_value, Price_in_thousands, Engine_size, Horsepower.

```python
My_Numerical_Columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size', 'Horsepower']

# Calculate variance for each numerical column
Variance_of_Columns = DF[My_Numerical_Columns].var()

# Calculate standard deviation for each numerical column
Std_deviation_of_Columns = DF[My_Numerical_Columns].std()

# Display the results
print("Variance for each column:")
print("-------------------------")
print(Variance_of_Columns)

print("\nStandard Deviation for each column:")
print("-------------------------------------")
print(Std_deviation_of_Columns)
```

## For the same columns Calculate IQR and identify outliers in these columns
Q1 = df[column].quantile(0.25)

outliers = df[(df[column] < Q1 - 1.5 * IQR) | (df[column] > Q3 + 1.5 * IQR)]

```python
import seaborn as sns

My_Numerical_Columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size', 'Horsepower']

for column in My_Numerical_Columns:   # Calculate IQR for each numerical column
    Q1 = DF[column].quantile(0.25)
    Q3 = DF[column].quantile(0.75)
    IQR = Q3 - Q1
    
    outliers = DF[(DF[column] < Q1 - 1.5 * IQR) | (DF[column] > Q3 + 1.5 * IQR)][column] # Identify outliers
    
    print(f"\nColumn: {column}")
    print(f"IQR: {IQR}")
    print(f"Number of Outliers: {len(outliers)}")
    print(f"Outliers: \n{outliers}")

    plt.figure(figsize=(10, 5)) # Visualize outliers using a box plot
    sns.boxplot(x=DF[column], color="Violet")
    plt.title(f'Box plot for {column}')
    plt.show()
```

## Make Histograms for all Numerical columns and identify the distribution of type by calculating
skewness;
Note : If the skewness is 0, it indicates a perfectly symmetrical distribution.
If the skewness is negative, it means the distribution is skewed to the left (long left tail).
If the skewness is positive, it means the distribution is skewed to the right (long right tail).

```python
My_Numerical_columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity', 'Power_perf_factor', 'Fuel_efficiency']

plt.figure(figsize=(15, 10)) # Plot histograms for numerical columns
DF[My_Numerical_columns].hist(bins=20, color="Violet", edgecolor='black', grid=False, layout=(4, 4), figsize=(15, 10))


skewness_values = DF[My_Numerical_columns].skew() # Calculate skewness for each numerical column

print("Skewness for each column:") # Display skewness values
print(skewness_values)

plt.tight_layout()
plt.show() # Show the plot
```
